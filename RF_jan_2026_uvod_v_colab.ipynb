## Gravitacijski pospešek

Splošna enačba s katero lahko izračunamo gravitacijski pospepešek ima obliko
$$
g(r) = \frac{G \cdot M}{r^2}.
$$

V tej enačbi je $G$ splošna gravitacijska konstanta, $M$ je masa objekta in $r$ razdalja med masnim središčem objekta in dano točko v prostoru.
