````python
class Planet:
  def __init__(self,pp:dict):
    self.masa = pp["masa"]
    self.polmer = pp["polmer"]*1000
    self.voda = pp["voda"]
    self.satelit = pp["sateliti"]
    self.T = pp["perioda"]
    self.ime = ''
    self.G = 6.67e-11
    self.g = self.gravitacija()

  def gravitacija(self):
    g = self.G*self.masa/(self.polmer**2)
    self.g = g
    return g

  def prosti_pad_iz_visine(self,h:float = 10):
    #Razdalje so ponade v metrih
    from math import sqrt
    t = sqrt(2*h/self.g)
    return t
````

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import Video

# Logistic map parameters
r = 3.8  # Control parameter
iterations = 200  # Number of iterations
x0 = 0.2  # Initial condition

# Compute the logistic map values
x_vals = [x0]
for _ in range(iterations):
    x_vals.append(r * x_vals[-1] * (1 - x_vals[-1]))

# Prepare the return map data
x_n = x_vals[:-1]  # x_n
x_np1 = x_vals[1:]  # x_n+1

# Setup for animation with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
fig.suptitle('Logistic Map Analysis', fontsize=16)

# --- Panel 1: Return Map ---
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
ax1.set_xlabel('$x_n$', fontsize=14)
ax1.set_ylabel('$x_{n+1}$', fontsize=14)
ax1.set_title('Return Map', fontsize=12)

# Plot the diagonal line (y = x) and logistic curve on ax1
x = np.linspace(0, 1, 500)
ax1.plot(x, x, 'gray', linestyle='--', label='y = x')  # Diagonal
ax1.plot(x, r * x * (1 - x), 'blue', label=rf'$x_{{n+1}} = r \cdot x_n (1 - x_n)$')
ax1.legend()

# Initialize the point plot and the lines for ax1
points_ax1, = ax1.plot([], [], 'ro', markersize=3, alpha=0.7)  # Red points
lines_ax1, = ax1.plot([], [], 'r-', linewidth=0.5, alpha=0.6)  # Red connecting lines

# --- Panel 2: Time Series ---
ax2.set_xlim(0, iterations)
ax2.set_ylim(0, 1)
ax2.set_xlabel('Iteration', fontsize=14)
ax2.set_ylabel('$x_n$', fontsize=14)
ax2.set_title('Time Series', fontsize=12)

# Initialize the line and current point for ax2
line_ax2, = ax2.plot([], [], 'g-', linewidth=1.0) # Green line for time series
point_ax2, = ax2.plot([], [], 'ro', markersize=5) # Red point for current iteration

# Animation function
def update(i):
    # Update points for Panel 1 (Return Map)
    points_ax1.set_data(x_n[:i+1], x_np1[:i+1])

    # Update connecting lines for Panel 1
    line_x = []
    line_y = []
    for j in range(i + 1):
        # Vertical line from (x_n[j], x_n[j]) to (x_n[j], x_np1[j])
        line_x.extend([x_n[j], x_n[j]])
        line_y.extend([x_n[j], x_np1[j]])

        # Horizontal line from (x_n[j], x_np1[j]) to (x_np1[j], x_np1[j])
        line_x.extend([x_n[j], x_np1[j]])
        line_y.extend([x_np1[j], x_np1[j]])
    lines_ax1.set_data(line_x, line_y)

    # Update line for Panel 2 (Time Series)
    line_ax2.set_data(range(i + 1), x_vals[:i + 1])

    # Update current point for Panel 2
    point_ax2.set_data([i], [x_vals[i]])

    return points_ax1, lines_ax1, line_ax2, point_ax2


# Create the animation
ani = FuncAnimation(fig, update, frames=len(x_n), interval=50, blit=False)

# Display the animation (optional in some environments, but crucial for Colab)
plt.close(fig) # Close the static plot to avoid displaying it before the video

# Save the animation as a video file
filename = "full_logistic_return_map_two_panels.mp4" # Changed filename to reflect two panels
ani.save(filename, fps=15, extra_args=['-vcodec', 'libx264'])

# Display the saved video
Video(filename)
````

### Primer: Matrično množenje s CPU (NumPy) in GPU (TensorFlow)

Ta program bo izvedel množenje velikih matrik z uporabo CPU (NumPy) in nato, če je na voljo, z uporabo GPU (TensorFlow), da prikaže razliko v hitrosti.

````python
import numpy as np
import time
import tensorflow as tf

# Preverimo, ali je GPU na voljo
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print(f"GPU je na voljo: {gpus}")
    using_gpu = True
else:
    print("GPU ni na voljo. Uporabljam samo CPU.")
    using_gpu = False

# Nastavitev velikosti matrik
matrix_size = 2_000 # Poskusi z večjo številko, npr. 5000, če imaš močan GPU

print(f"\nIzvajanje matričnega množenja za matrike velikosti {matrix_size}x{matrix_size}")

# --- Matrično množenje na CPU (NumPy) ---
print("\nTestiranje na CPU (NumPy)...")
A_cpu = np.random.rand(matrix_size, matrix_size).astype(np.float32)
B_cpu = np.random.rand(matrix_size, matrix_size).astype(np.float32)

start_time = time.time()
C_cpu = np.dot(A_cpu, B_cpu)
end_time = time.time()

print(f"Čas izvajanja na CPU: {end_time - start_time:.4f} sekund")

# --- Matrično množenje na GPU (TensorFlow), če je na voljo ---
if using_gpu:
    print("\nTestiranje na GPU (TensorFlow)...")

    # TensorFlow operacije so privzeto izvedene na GPU, če je na voljo
    with tf.device('/GPU:0'): # Eksplicitno določimo GPU, če jih je več
        A_gpu = tf.random.uniform([matrix_size, matrix_size], dtype=tf.float32)
        B_gpu = tf.random.uniform([matrix_size, matrix_size], dtype=tf.float32)

        start_time = time.time()
        C_gpu = tf.matmul(A_gpu, B_gpu)
        # .numpy() prisili izračun in prenos rezultata na CPU, zato merimo do sem
        C_gpu.numpy()
        end_time = time.time()

    print(f"Čas izvajanja na GPU: {end_time - start_time:.4f} sekund")
else:
    print("\nPreskočim testiranje na GPU, ker ni na voljo.")
````
